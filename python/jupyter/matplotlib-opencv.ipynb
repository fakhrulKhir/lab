```python
# install the packages if you have not
# !pip install cv2 matplotlib

import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('images/bali.jpg')
print('type of img:', type(img))
```

```python
# convert to gray
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

# uncomment to show the image in cv2 (image will be displayed in new window)
# cv2.imshow('Bali in gray', gray_img)
# cv2.waitKey(0) # wait forever for user to press any key

# show the bgr image using matplotlib in gray mode
plt.imshow(imgray, cmap='gray')
plt.show()

```

```python
# convert cv2 BGR to RGB because cv2 uses BGR to display image and matplotlib uses RGB
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
print('- rows in `im_rgb`:', len(im_rgb))

print('- display all rows in plot:')
plt.plot(im_rgb[500]) # plot the image channel on row 500
plt.show()
```

```python
# cropping the rgb image
im_crop = im_rgb[400:900,400:700] # take row 400 to 900, and column 400 to 700
plt.imshow(im_crop)
plt.show()
```

```python
# display image in single channel (blue)
im_blue = im_rgb[:,:,2] # r=0, g=1, b=2
plt.imshow(im_blue) # brighter area indicates higher blue values
plt.show()

plt.plot(im_blue[255,:]) # load and plot value of columns at row 255 of `im_blue`
plt.show()
```
