```python
import visdom
import numpy as np
import time

v = visdom.Visdom()

assert v.check_connection()
print('visdom is up')
```

```python
class VisdomLine():
    def __init__(self, visdom, win, start_x, start_y):
        """line graph that follows"""
        
        self.v = visdom
        self.win = win
        
        self.start_x = start_x
        self.start_y = start_y
    
    def append_y(self, new_y, name, x_increment = 1):
        latest_x = self.start_x
        self.start_x += x_increment
        
        latest_y = self.start_y
        self.start_y = new_y
        
#       print('latest', self.start_x, new_y)
        y_transition = np.linspace(latest_y, self.start_y, 10)
        x_transition = np.linspace(latest_x, self.start_x, 10)

        self.v.line(
            X=x_transition,
            Y=y_transition,
            win=self.win,
            name=name,
            update='append',
        )
```

```python
# simulate change of moods:
# -1 = sad
#  0 = happy
#  1 = angry

win = v.line(X=np.array([0, 0, 0]), 
             Y=np.array([-1, 0, 1]), 
             opts=dict(
                 title='EMOS: Emotion of speech',
            ))
line = VisdomLine(v, win, 0, 0)

moods_spec = {'happy': 0, 'sad':-1, 'angry':1}
moods = ['happy', 'happy', 'happy', 'sad', 'sad', 'angry', 'angry', 'happy', 'sad', 'angry', 'happy', 'sad', 'sad', 'sad']

for mood in moods:
    line.append_y(moods_spec[mood], 'mood')
    time.sleep(0.5)
```
