```python
# import the package
# hint (macOS): hover on variable and press shift+tab to display help (docstring). 
#               cmd+shif+p to search all commands.

# uncomment the code below to install 
# matplotlib package if you have not.
# in Jupyter notebook, the `!` is 
# used to execute shell command.
#
# !pip install matplotlib

import matplotlib.pyplot as plt
```

```python
# creating a histogram 
# hint: press ctrl+enter on macOS to execute this block
val = [0, 0.5, 0.1, 0.4, 4, 2, 4, 5, 5, 1, 2]

# use `hist` function to create histogram
plt.hist(val)
    
# show the histogram
plt.show()
```

```python
# creating a plot
x = [1,2,4,5,5,6]
y = [1,1,2,3,4,5]

# use `plot` function to create a plot
plt.plot(x, y)

plt.show()
```
